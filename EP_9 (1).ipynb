Así, calculando el vector de Poyting

```python
#Le damos formato de vectores a E y B
Ev = [E,0,0]
Bv = [0,B,0]
Ev_1 = [E_1,0,0]
Bv_1 = [0,B_1,0]
```

```python
#Función de producto cruz
def proCr(a, b):
        c = [a[1]*b[2] - a[2]*b[1],
            a[2]*b[0] - a[0]*b[2],
            a[0]*b[1] - a[1]*b[0]]
        return c
```

```python
#Calculando númericamente A el vector de Poyting para los mismos intervalos del inciso a)


Poyt = proCr(Ev,Bv)

Poyt_1 = proCr(Ev_1,Bv_1)
```

# d)

Finalmente calcularemos sus promedios de estos vectores finalmente definiendo a \textbf{g} como

\begin{equation}
\textbf{g} = \frac{1}{c}\epsilon_0 E_0^2cos^2(kz - \omega t+ \delta) \hat{z} = \frac{1}{c}u \hat{z}
\end{equation}


Por lo que hemos terminado de definir los otros 2 vectores en terminos de u, con eso calcularemos la integral (1) con el siguiente algoritmo:

```python
#Funciones que hacen el producto cruz y generan una función para integrar, en este caso específico calculamos

suma = 0    
    
def f(x):
    return pow(mt.cos(k_0*1-2*mt.pi*x),2)


a_i = 0
b_i = 1
N = 10000
dx = (b_i - a_i)/N
c_i = a_i + dx/2.0 
```

```python
##Algoritmo de integración por sumas de Riemman

for i in range(0,N):
    x_i = c_i + (i*dx)
    suma = suma + f(x_i)

intFinal = suma*dx;
print("La integral de la ecuación (1)es :",intFinal)
```

Que era la integral que se había calculado al inicio de manera conceptual y aquí de tal manera vemos que esta integral es (1)=\frac{1}{2}
Veamos ahora entonces que los promedios serán:
\begin{equation}
\langle u \rangle = \frac{1}{2} \epsilon_0 E_0^2
\end{equation}
\begin{equation}
\langle S \rangle = \frac{1}{2}c \epsilon_0 E_0^2 \hat{z}
\end{equation}
\begin{equation}
\langle g \rangle = \frac{1}{2c} \epsilon_0 E_0^2 \hat{z}
\end{equation}


Numéricamente tendremos finalmente y usando el valor calculado de la integral de (1)

```python
u_p = intFinal*e_0*pow(E_0,2)
S_p = intFinal*c_0*e_0*pow(E_0,2)
g_p = intFinal*(1/c_0)*e_0*pow(E_0,2)
```

```python
print("El promedio del vector u es: ")
print(u_p)
print("El promedio del vector S en la componente z es: ")
print(S_p, "z")
print("El promedio del vector g en la componente z es: ")
print(g_p, "z")
```

En 3d Graficando con z, t como varibles independientes y S la parte dependiente

```python
from mpl_toolkits.mplot3d import Axes3D

import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np


fig = plt.figure()
ax = fig.gca(projection='3d')

# Make data.
t_00 = np.linspace(0,1.5,100)
z_00 = np.linspace(0,1.5,100)
t_00, z_00 = np.meshgrid(t_00, z_00)
t = k_0*z_00-om_0*t_00
Z = (E_0*np.cos(t))*((1/c_0)*E_0*np.cos(t))

# Plot de la superficie
surf = ax.plot_surface(t_00, z_00, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)


ax.set_zlim(-0.0000001, 0.0000002)
ax.zaxis.set_major_locator(LinearLocator(2))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Agregando colores
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
