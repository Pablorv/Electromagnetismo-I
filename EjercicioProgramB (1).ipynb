```python
import math
import numpy as np
from matplotlib import pyplot as plt
import sympy as sp
```

```python
sR = [0.2,0.1, 0.05,0.02,0.01]
f = [1.286,1.167, 1.094,1.042,1.023]
```

```python

plt.scatter(sR,f)
plt.ion()
plt.xlabel("s/R")
plt.ylabel("f")
plt.title("Representacion gráfica")
plt.scatter(sR, f, color="blue", linewidth=5, linestyle="-", label="")
plt.legend(loc='upper right')
```

```python
plt.plot(sR, f)
plt.ion()
plt.xlabel("s/R")
plt.ylabel("f")
plt.title("Representacion gráfica")
plt.plot(sR, f, color="yellow", linewidth=5, linestyle=":", label="-")
plt.legend(loc='upper right')
```
