Buscamos una represtación númerica de la siguiente expresión así mismo como que esta misma representación nos ayude a encontrar los datos que dejen entrever por medio de gráficas como estará comportando la función.
\\
Así definimos entonces $Z$ como
\\
\begin{equation}
Z = i \omega L+\frac{1}{\frac{1}{R}+i \omega C}
\end{equation}
\\
De lo anterior podemos notar primero que se hará una representación general de Z donde para Z_1 veremos como la haremos fija para R,C y L, con la intención de visualizar como se puede operar de manera sencilla. 
\\
Luego, para Z, haremos su representación imaginaria dejando fijos tanto R como C y variando L, luego gráficaremos a la parte real como la parte imaginaria en gráficas separadas y con esto buscaremos una interpretación Física.

```python
import numpy as np
import math
from matplotlib import pyplot as plt
import sympy as sp
```

##### Para valores fijos demostrando que se puede calcular númericamente operaciones con números complejos obteniendo una parte real y una imaginaria. 

```python
L_1 = 3
R_1 = 1
C_1 = 0.004
w_1 = 1/pow(L_1*C_1,1/2)
```

```python
Z_1 = (w_1*L_1)*1j+(1/((1/R_1)+(w_1*C_1)*1j))
print("Representación de este cálculo con su parte real e imaginaria\n", Z_1)
```

```python
a=[0]
b=[0]
```

```python
L = np.linspace(0.0000001,2,1000)
R = 1
C = 0.004
w = 1/pow(L*C,1/2)
```

```python
Z = (w*L)*1j+(1/((1/R)+(w*C)*1j))
print("Así, obtenemos los siquiente valores para Z\n", Z)
```

```python
a = Z.real
b = Z.imag
c = np.linspace(0,10,1000)
```

```python
print("Separando en su parte real\n", a)
```

```python
print("Separando en su parte Imaginaria\n", b)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def move_spines():
    fix, ax = plt.subplots()
    for spine in ["left", "bottom"]:
        ax.spines[spine].set_position("zero")
    
    for spine in ["right", "top"]:
        ax.spines[spine].set_color("none")
    
    return ax

ax = move_spines()
ax.set_xlim(-2, 2)
ax.set_ylim(-5, 5)
ax.grid()
ax.scatter(a, b)
plt.title("Plano de Argand")
plt.show()
```

```python
plt.plot(c, a)
#plt.plot(c, b)
#plt.plot(c, Z)
plt.ion()
plt.xlabel("Variación de L")
plt.ylabel("Valores en Re")
plt.title("Impedancia")
plt.plot(c, a, color="blue",  linewidth=2.5, linestyle="-", label="Parte Real")
#plt.plot(c, b, color="red",  linewidth=2.5, linestyle="-", label="Parte Imaginaria")
#plt.plot(c, Z, color="red",  linewidth=2.5, linestyle="-", label="Parte Imaginaria")
plt.legend(loc='upper right')
```
